## Web Based Loader -- Scrapes raw HTML and extracts text

```python
from langchain_community.document_loaders import WebBaseLoader
## For loading entire web
loader = WebBaseLoader(web_path="https://en.wikipedia.org/wiki/Naruto") # we can use "web_paths" for multiple webs
loader.load()
```

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4 # bs4 -- Beatifull_Soup used for web scrapping tasks'
## for loading selected parts of web
loader=WebBaseLoader(web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
                     bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                         class_=("post-title","post-content","post-header")
                     ))
                     )
loader.load()
```

## Arxiv Loader--used to load research papers from arxiv

arXiv (pronounced like "archive") is a free, open-access repository of scholarly research papers. It is widely used in fields such as:

Physics

Mathematics

Computer Science

Statistics

Electrical Engineering and Systems Science

Quantitative Biology

Economics

```python
from langchain_community.document_loaders import ArxivLoader

loader = ArxivLoader(query="1706.03762", load_max_docs = 2) # query=1706.03762--- it is the search orload parameter which is the research paper no.
loader.load()
```

## Wikipedia Loader--used to load content from wikipedia

```python
from langchain_community.document_loaders import WikipediaLoader

loader = WikipediaLoader(query="Hunter x Hunter",load_max_docs=2)
loader.load()
```
